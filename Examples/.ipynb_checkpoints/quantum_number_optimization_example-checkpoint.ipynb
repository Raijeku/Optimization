# Quantum random optimization
This optimization method uses random number generation on a quantum computer to optimize a function.

```python
import sys
sys.path.append("..")

from quantum_random_optimization import *
import numpy as np
```

Here we generate 1000 random numbers using the ibmq_qasm_simulator machine.

```python
numbers = np.array(generate_random(1000,'ibmq_qasm_simulator'))
numbers
```

```python
print('The {0} generated numbers have a mean of {1} and a standard deviation of {2}'.format(numbers.size,numbers.mean(),numbers.std()))
```

We now perform an optimization of the $-x^2 $ function.

```python
results = quantum_random_optimization(-3,2,1000,'-x**2','ibmq_qasm_simulator')
results
```

```python
print('The best value of x is {0}, where f(x) = {1}'.format(results.iloc[-1]['max_x'],results.iloc[-1]['max_f(x)']))
```
